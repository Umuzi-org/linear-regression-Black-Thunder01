## Updated Salary assignment

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
import pylab as py
import numpy as np
from scipy import stats
import statsmodels as sm
from mpl_toolkits import mplot3d
```

```python
data = pd.read_csv('salary.csv')
```

```python
np.where(pd.isnull(data))
```

```python
data = data.dropna()
```

```python
data.head()
```

#### Extracting my Data
------------
* Since the data contains data contains data for different business fields I'm going to seperate the data according to their respective fields.

```python
#  field
engineering = data[data['Field'] == 1]
Finance = data[data['Field'] == 2]
HR = data[data['Field'] == 3]
marketing = data[data['Field'] == 4]
```

```python
# Independent variables/ yearsworked for field of work
xE = np.array(engineering.yearsworked)
xF= np.array(Finance.yearsworked)
xH= np.array(HR.yearsworked)
xM= np.array(marketing.yearsworked)

# Dependent variables/ Salaries for field of work
yE = np.array(engineering.salary)
yF= np.array(Finance.salary)
yH= np.array(HR.salary)
yM= np.array(marketing.salary)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([1, 0.5, 1, 0.5])
ax2 = fig.add_axes([1, 0, 1, 0.5])
ax3 = fig.add_axes([1, 1, 1, 0.5])
ax4 = fig.add_axes([1, 1.5, 1, 0.5])

ax1.scatter(xE,yE, color="blue")
ax2.scatter(xF,yF, color="red")
ax3.scatter(xH,yH, color="orange")
ax4.scatter(xM,yM, color="black");
```

```python
# r value for Engineering
EM, Einter, Er, Ep, Estd_err = stats.linregress(xE,yE)
print("R value for Engineering : ", Er)
print("")
# r value for Financing
FM, Finter, Fr, Fp, Fstd_err = stats.linregress(xF,yF)
print("R value for Finance :", Fr)
print("")
# r value for Human Resources
HM, Hinter, Hr, Hp, Hstd_err = stats.linregress(xH,yH)
print("R value for HR :", Hr)
print("")
# r value for Marketing
MM, Minter, Mr, Mp, Mstd_err = stats.linregress(xM,yM)
print("R value for Marketing", Mr)
# Average r value of the fields
print("Average r-value among all Field :",np.mean([Er,Fr,Hr,Mr]))
```
